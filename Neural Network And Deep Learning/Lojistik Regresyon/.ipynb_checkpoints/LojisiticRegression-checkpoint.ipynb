```python
#simdi kullanilabilirlik acisindan training setleri ((64*64*3),209) seklinde bir hale getirecegiz...
#train setin transpozunu aldÄ±ÄÄ±mÄ±z anda satÄ±r sayÄ±sÄ± deÄiÅken bir deÄer olmadÄ±ÄÄ± iÃ§in bu Ã¼Ã§ deÄer Ã§arpÄ±lÄ±r 
#ve deÄerler alt alta dizilir. SatÄ±r sayÄ±sÄ±da sÃ¼tun sayÄ±sÄ± olur.
#AynÄ±sÄ± test iÃ§inde geÃ§erlidir.
train_set_x_flatten = train_set_x_orig.reshape(train_set_x_orig.shape[0], -1).T
test_set_x_flatten = test_set_x_orig.reshape(test_set_x_orig.shape[0], -1).T

print("Transpoze iÅleminden sonra X train matrisi : (" + str(train_set_x_flatten.shape[0]) + "," + str(train_set_x_flatten.shape[1]) + ")")
print("Transpoze iÅleminden sonra X test matrisi : (" + str(test_set_x_flatten.shape[0]) + "," + str(test_set_x_flatten.shape[1]) + ")")
```

```python
#verileri standartlastÄ±r...
train_set_x = train_set_x_flatten/255.
test_set_x = test_set_x_flatten/255.

print(str(train_set_x.shape[0]) +","+str(train_set_x.shape[1]))
```

```python
#gerekli ogrenme parametrelerini 0 ile baslat...
def initialize_with_zeros(dim):
    w = np.zeros(shape=(dim,1))
    b = 0
    #baslattÄ±gÄ±mÄ±z degerlerin istediÄimiz boyutlara sahip olup olmadÄ±larÄ±nÄ± kontrol et...
    assert(w.shape==(dim,1))
    assert(isinstance(b,float) or isinstance(b,int))
    return w, b
```

```python
#sigmoid fonksiyonunu tanimliyoruz....
def sigmoid(z):
    s = 1/(1+np.exp(-z))
    return s
```

```python
#ileri ve geri yayÄ±lma....
def propagate(w,b,X,Y):
    #ileri yayÄ±lma...
    m=X.shape[1]
    A = sigmoid(np.dot(w.T,X)+b)
    cost = (-1/m) * np.sum(Y * np.log(A) + (1-Y) * np.log(1-A))
    
    #geri yayÄ±lma....
    dw = (1/m) * np.dot(X, (A - Y).T)
    db = (1/m) * np.sum(A - Y)
    
    grads = {"dw" : dw,
             "db" : db}
    
    return grads,cost
```

```python
#geri yayÄ±lmayÄ± gradient descent ile bir iterasyona sokarak optimizasyon islemi yapma...
def optimization(w, b, X, Y, num_iteration, learning_rate, print_cost=False):
    costs = []
    for i in range(num_iteration):
        
        grads,cost = propagate(w,b,X,Y)
        dw=grads["dw"]
        db=grads["db"]
        
        w = w - (learning_rate * dw)
        b = b - (learning_rate * db)
        
        if(i%100==0):
            costs.append(cost)
            
        if(print_cost and i%100==0):
            print ("%i Ä°terasyondaki cost : %f" % (i, cost))
            
    params = {"w" : w,
              "b" : b}
        
    grads = {"w" : dw,
             "b" : db}
        
    return params, grads, costs
```

```python
#GÃ¶nderilen veri setinin Y tahminlerini yapar. BÃ¶ylece veri setindeki verilerin tahmini yapÄ±lmÄ±Å olur.....
def predict(w, b, X):
    Y_prediction = np.zeros(shape = (1, X.shape[1]))
    w = w.reshape(X.shape[0], 1)
    
    A = sigmoid(np.dot(w.T,X)+b)
    
    for i in range(A.shape[1]):
        if(A[0,i]>0.5):
            Y_prediction[0,i] = 1
        else:
            Y_prediction[0,i] = 0
            
    assert(Y_prediction.shape == (1, X.shape[1]))
    return Y_prediction
```

```python
#Logistik Regresyonun UygulanmasÄ±....
def logisticRegressionModel(X_train, Y_train, X_test, Y_test, num_iteration, learning_rate, print_cost=False):
    w, b = initialize_with_zeros(X_train.shape[0])
    
    params, grads, costs = optimization(w, b, X_train, Y_train, num_iteration, learning_rate, print_cost)
    
    w = params["w"]
    b = params["b"]
    
    Y_prediction_test = predict(w,b,X_test)
    Y_prediction_train = predict(w,b,X_train)
    
        #aritmetik ortalamasÄ±nÄ± alacaÄÄ±z ve 100 ile Ã§arparak yÃ¼zdesini bulacaÄÄ±z...
    print("Train'in doÄruluÄu : {} %".format(np.abs(100 - np.mean(np.abs(Y_prediction_train - Y_train)) * 100)))
    print("Test'in doÄruluÄu : {} %".format(np.abs(100 - np.mean(np.abs(Y_prediction_test - Y_test))*100)))
    
    
    d = {"costs" : costs,
         "Y_prediction_test" : Y_prediction_test, 
         "Y_prediction_train" : Y_prediction_train, 
         "w" : w, 
         "b" : b, 
         "learning_rate" : learning_rate,
         "num_iterations": num_iteration}
    
    return d
    
```

```python
d = logisticRegressionModel(train_set_x, train_set_y, test_set_x, test_set_y, num_iteration=2000, learning_rate=0.005, print_cost = True)
```

```python
#test seti iÃ§inden bir resim seÃ§iyoruz ve gÃ¶zlemliyoruz...
index = 5
plt.imshow(test_set_x[:,index].reshape((num_px, num_px, 3)))
print ("y = " + str(test_set_y[0, index]) + ", you predicted that it is a \"" + classes[d["Y_prediction_test"][0, index]].decode("utf-8") +  "\" picture.")
```

```python
#denetimli Ã¶Ärenme eÄrisi...
costs = np.squeeze(d["costs"])
plt.plot(costs)
plt.xlabel("Iterasyonlar")
plt.title("Learning Rate = " + str(d["learning_rate"]))
plt.show
```

```python
#Ã¶Ärenme katsayÄ±larÄ±nÄ±n ne tÃ¼r bir deÄiÅime sebep olduklarÄ±nÄ± gÃ¶rdÃ¼k... 
learning_rates = [0.003, 0.001, 0.005]
models = {}
for i in learning_rates:
    print ("learning rate is: " + str(i))
    models[str(i)] = logisticRegressionModel(train_set_x, train_set_y, test_set_x, test_set_y, num_iteration = 2000, learning_rate = i, print_cost = False)
    print ('\n' + "-------------------------------------------------------" + '\n')

for i in learning_rates:
    plt.plot(np.squeeze(models[str(i)]["costs"]), label= str(models[str(i)]["learning_rate"]))

plt.ylabel('cost')
plt.xlabel('iterations')

legend = plt.legend(loc='upper center', shadow=True)
frame = legend.get_frame()
frame.set_facecolor('0.90')
plt.show()
```
